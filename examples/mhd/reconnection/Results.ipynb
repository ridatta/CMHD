# Output at given time

```python
t = [2000,4000,6000]
field = 'p'

fig, axs = plt.subplots(len(t),1,sharex=True,dpi=300)
for ii in range(len(t)):
    data = np.loadtxt('./output/' + field + '-' + str(t[ii]) + '.txt')
    Bx = np.loadtxt('./output/' + 'Bx' + '-' + str(t[ii]) + '.txt')
    By = np.loadtxt('./output/' + 'By' + '-' + str(t[ii]) + '.txt')
    im = axs[ii].imshow(data)
    axs[ii].set_title('t = ' + str(t[ii]))
    m, n = np.shape(Bx)
    xx, yy = np.meshgrid(np.arange(0,n),np.arange(0,m))
    axs[ii].streamplot(xx,yy,Bx,By,linewidth=0.5,color='w',density=1.5,arrowsize=0.25,broken_streamlines=True)
    
fig.tight_layout(pad=0.1)
# plt.colorbar(im)
```
