# Comapre w/ Analytical Result

```python
data = np.loadtxt("./output/rho.txt")
m,n = data.shape
rho = data[m//2,n//2:]

r = np.linspace(0,0.5,n//2)

fig, axs = plt.subplots(3,1,dpi=200,figsize=(4,6),sharex=True)
axs[0].plot(r,rho)
axs[0].set_ylabel('$\\rho$')
axs[0].set_yscale('log')
axs[0].set_xlim([1e-1,7])

data = np.loadtxt("./output/p.txt")
p = data[m//2,n//2:]

axs[1].plot(r,p)
axs[1].set_ylabel('p')

data = np.loadtxt("./output/u.txt")
u = data[m//2,n//2:]

axs[2].plot(r,u)
axs[2].set_ylabel('u')

axs[-1].set_xlabel('r')

for ii in range(axs.shape[0]):
    axs[ii].grid()
    axs[ii].set_xlim([0,0.5])


```
