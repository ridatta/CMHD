```python

u = np.loadtxt("./output/u.txt")
dU = np.gradient(u,1/u.shape[0],axis=1)
p = np.loadtxt("./output/p.txt")
rho = np.loadtxt("./output/rho.txt")
gamma = 5/3

T = p / rho
beta = 0.5
alpha = 2
Qrad = 0.5* rho**alpha * T**beta

Qcomp = -gamma *  p * dU / (gamma-1)

plt.imshow(Qcomp,cmap='bwr')
plt.colorbar()
```

# Output at given time

```python
t = np.arange(0,10001,2000)
field = 'rho'

fig, axs = plt.subplots(t.shape[0],1,sharex=True,dpi=200)
for ii in range(len(t)):
    data = np.loadtxt('./output/' + field + '-' + str(t[ii]) + '.txt')
    im = axs[ii].imshow(data,cmap='viridis')
    axs[ii].set_title('t = ' + str(t[ii]))
# plt.colorbar(im,orientation="horizontal")
```
