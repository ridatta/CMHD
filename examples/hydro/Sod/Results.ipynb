# Comapre w/ Analytical Result

```python
data = np.loadtxt("./output/rho.txt")
rho_mean = np.mean(data,0)
x = np.linspace(0,1,rho_mean.shape[0])

plt.plot(x,rho_mean,'o',label='simulation')
plt.grid()
plt.xlabel('x')
plt.ylabel('$\\rho$')
plt.xlim([0,1])

data = np.loadtxt('sod_0p2.csv',delimiter=',')
plt.plot(data[:,0],data[:,1],label='analytical')

plt.legend()
```
