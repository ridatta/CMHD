# Final Output

```python
field = 'rho'
t = ''
data = np.loadtxt('./output/' + field + t + '.txt')
plt.imshow(data,origin='lower',cmap='inferno')
plt.colorbar()


v = np.loadtxt('./output/v' + t +'.txt')
u = np.loadtxt('./output/u' + t +'.txt')
m,n = u.shape
x, y = np.arange(0,n,1), np.arange(0,m,1)
plt.streamplot(x,y,u,v, density=0.4, color='tab:green',broken_streamlines=False)
plt.ylim(ymin=0)
```

```python
ny, nx = data.shape

xs, ys = [], []

for ii in range(150,200,(ny-20)//50):
    
    rho_out = data[ii,:]
    plt.plot(rho_out,label='y='+str(ii))
    
    ys.append(ii)
    xs.append(np.argmax(rho_out))
    
    

plt.grid()
plt.legend()
plt.xlabel('x [-]')
plt.ylabel('$\\rho$ [-]')

xs = np.asarray(xs)
ys = np.asarray(ys)
```

```python
plt.plot(xs,ys,'o')
plt.xlabel('$x_s$')
plt.ylabel('$y_s$')

from scipy.optimize import curve_fit

def fn(x,m,c):
    return m*x + c

popt, pcov = curve_fit(fn, xs, ys)
plt.plot(xs, fn(xs, *popt), 'r-')

angle = np.arctan(popt[0])
print('shock angle = %1.3f deg'%(np.rad2deg(angle)))
```
