```python
field = 'u'
data = np.loadtxt('./output/' + field + '.txt')
plt.imshow(data,origin='lower',cmap='bwr',vmin=-1,vmax=1)
plt.colorbar()
```

```python
rho = np.loadtxt("./output/rho.txt")
v = np.loadtxt("./output/v.txt")
u = np.loadtxt("./output/u.txt")
bnx = np.loadtxt("./output/bnx-0.txt")
bny = np.loadtxt("./output/bny-0.txt")

m, n = np.shape(rho)
out = v[:,60]

plt.plot(out,'o')
plt.grid()


```

```python

U1 = rho * u
U2 = rho * v
vn = U1 * bnx + U2 * bny
U1 = U1 - 2 * vn * bnx
U2 = U2 - 2 * vn * bny


fig, axs = plt.subplots(6,1,sharex=True,figsize=(4,6),dpi=150)
axs[0].plot(rho,'-o',label='rho')
axs[1].plot(u,'-o',label='u')
axs[2].plot(v,'-o',label='v')
axs[3].plot(bnx,'-o',label='bnx')
axs[3].plot(bny,'-o',label='bny')
axs[4].plot(rho*u,'-o',label='$\\rho u$')
axs[4].plot(U1,'-o',label='$U_1$')

axs[5].plot(rho*v,'-o',label='$\\rho v$')
axs[5].plot(U2,'-o',label='$U_2$')

for ii in range(axs.shape[0]):
    axs[ii].grid()
    axs[ii].legend()
```
